```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.title("My plot")
plt.show()
```

If we provide a single list or array to the plot() command, matplotlib assumes it as a sequence of y-values,
and automatically generates x-values.

Since python range starts with 0, the default x vector has same length as y but starts with 0.
Hence x-data are [0,1,2,3]

#### Plot x vs y

```python
plt.plot([1,2,3,4],[1,4,8,16]) #x-values, y-values
plt.xlabel("Numbers")
plt.ylabel("Square")
plt.title("My plot")
plt.grid() # grid on
plt.show()
```

For every x,y pair of arguments there is an optional third argument, called 'format string', 
which indicates color and line type of the plot

```python
plt.plot([1,2,3,4],[1,4,8,16], 'ro') #x-values, y-values, r for red and o of oceanic dot
plt.xlabel("Numbers")
plt.ylabel("Square")
plt.title("My plot")
plt.grid() # grid on
plt.show()
```

```python
import numpy as np
a = np.arange(0,5,0.2)
plt.plot(a,a**2, 'b--', label='^2') 
plt.plot(a,a**2.2, 'rs', label='^2.2')
plt.plot(a,a**2.5, 'b^', label='^2.5')
plt.xlabel("Numbers")
plt.ylabel("power")
plt.title("My plot")
plt.legend()
plt.grid() # grid on
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,8,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1,y1,x2,y2)
#use keyword args
plt.setp(lines[0],color='r',linewidth=0.5)

#matlab style string value pair
plt.setp(lines[1],'color','g', 'linewidth','1')
plt.grid()
plt.plot()
```

# Working with Multiple Figures and axes

```python
x1 = [1,2,3,4]
y1 = [1,4,8,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

plt.figure(1)
# the subplot() command specifies numrows, numcols
# figurenum where num ranges from 1 to numrows*numcols

plt.subplot(321) #number specifies: no. of rows, no. of cols and subfigure number
plt.grid()
plt.plot(x1,y1,'b-')

plt.subplot(212) #number specifies: no. of rows, no. of cols and subfigure number
plt.grid()
plt.plot(x2,y2,'r--')

plt.figure(1)
# the subplot() command specifies numrows, numcols
# figurenum where num ranges from 1 to numrows*numcols

plt.subplot(321) #number specifies: no. of rows, no. of cols and subfigure number
plt.grid()
plt.plot(x1,y1,'b-')

plt.subplot(212) #number specifies: no. of rows, no. of cols and subfigure number
plt.grid()
plt.plot(x2,y2,'r--')
```
