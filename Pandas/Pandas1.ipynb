## Concatenate DataFrames

```python
Indian_weather = pd.DataFrame({
    "city" : ['Delhi','Chamoli','Banglore'],
    'temperature' : [32,25,30],
    'humidity' : [80,60,78]
})
Indian_weather
```

```python
Uttarakhand_weather = pd.DataFrame({
    "city" : ['Dehradun','Chamoli','Haridwar'],
    'temperature' : [32,25,30],
    'humidity' : [80,60,78]
})
Uttarakhand_weather
```

```python
# concatenate two dataframes
df = pd.concat([Uttarakhand_weather,Indian_weather])
df
```

```python
# ignore index
# concatenate two dataframes
df = pd.concat([Uttarakhand_weather,Indian_weather], ignore_index=True)
df
```

```python
# concatenate two dataframes column wise
df = pd.concat([Uttarakhand_weather,Indian_weather], axis=1)
df
```

## Merge DataFrames

```python
temp_df = pd.DataFrame({
    "city" : ['Delhi','Chamoli','Banglore','Chandigarh'],
    'temperature' : [32,25,30,25]
})
temp_df
```

```python
humidity_df = pd.DataFrame({
    "city" : ['Delhi','Banglore','Chamoli'],
    'humidity' : [80,60,78]
})
humidity_df
```

```python
#merge is like a join command in SQL
# merge two dataframes without explicitliy mention index
df = pd.merge(temp_df,humidity_df, on='city')
df
```

```python
df = pd.merge(temp_df,humidity_df, on='city', how='outer')
df
```

# Numerical Indexing (.loc vs iloc)

```python
df = pd.DataFrame([1,2,3,4,5,6,8,7,9])
df
```

Default Indexing

```python
df = pd.DataFrame([1,2,3,4,5,6,8,7,9] , index=[11,12,11,13,14,15,1,2,3])
df
```

Custom Indexing

```python
df.loc[1] #gives value according to custom index
```

```python
df.iloc[1] # gives value according to the row number or original index
```
